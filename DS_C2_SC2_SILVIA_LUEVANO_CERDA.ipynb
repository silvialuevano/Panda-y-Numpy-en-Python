### a. Extraer la información del archivo.

```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame de pandas
country_vaccinations = pd.read_csv('https://raw.githubusercontent.com/silvialuevano/Panda-y-Numpy-en-Python/main/country_vaccinations.csv')

# Mostrar las primeras 5 filas del DataFrame
display(country_vaccinations.head())
```



---


b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
print("Información inicial del DataFrame (estructura y tipos de datos):")
country_vaccinations.info()

# Convertir la columna 'date' a tipo datetime64
country_vaccinations['date'] = pd.to_datetime(country_vaccinations['date'])

print("\nInformación del DataFrame después de convertir 'date' a datetime64:")
country_vaccinations.info()
```


---
c. Determinar la cantidad de vacunas aplicadas de
cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
pd.set_option('display.float_format', '{:,.0f}'.format)
total_vaccinations_by_company = country_vaccinations.groupby('vaccines')['total_vaccinations'].sum()
display(total_vaccinations_by_company)
pd.reset_option('display.float_format')
```



---

d. Obtener la cantidad de vacunas aplicadas en todo el mundo.




```python
pd.set_option('display.float_format', '{:,.0f}'.format)
total_vaccinations_worldwide = country_vaccinations['total_vaccinations'].sum()
print(f"Total de vacunas aplicadas en todo el mundo: {total_vaccinations_worldwide:,.0f}")
pd.reset_option('display.float_format')
```



---

e. Calcular el promedio de vacunas aplicadas por país.


---



```python
pd.set_option('display.float_format', '{:,.0f}'.format)
average_daily_vaccinations_per_country = country_vaccinations.groupby('country')['daily_vaccinations'].mean()
display(average_daily_vaccinations_per_country)
pd.reset_option('display.float_format')
```



---

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo


---



```python
pd.set_option('display.float_format', '{:,.0f}'.format)
date_to_filter = '2021-01-29'
vaccinations_on_date = country_vaccinations[country_vaccinations['date'] == date_to_filter]['daily_vaccinations'].sum()
print(f"Cantidad de vacunas aplicadas el {date_to_filter} en todo el mundo: {vaccinations_on_date:,.0f}")
pd.reset_option('display.float_format')
```



---
g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.


---




```python
conDiferencias = country_vaccinations.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
display(conDiferencias.head())
display(conDiferencias.iloc[89:110])
diferencias_no_cero = conDiferencias[conDiferencias['diferencias'] != 0]
display(diferencias_no_cero)
```



---


h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.


---



```python
country_vaccinations_sorted = country_vaccinations.sort_values(by='date')

print("Primer registro (ordenado por fecha):")
display(country_vaccinations_sorted['date'].head(1))

print("\nÚltimo registro (ordenado por fecha):")
display(country_vaccinations_sorted['date'].tail(1))

time_period = country_vaccinations_sorted['date'].tail(1).iloc[0] - country_vaccinations_sorted['date'].head(1).iloc[0]

total_days = time_period.days

print(f"Total de días: {total_days}")

years = total_days // 365
remaining_days_after_years = total_days % 365
months = remaining_days_after_years // 30  # Approximate months
days = remaining_days_after_years % 30

print(f"Periodo de tiempo entre registros: {years} año, {months} meses y {days} días")
```



---

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).


---



```python
conCantidad = country_vaccinations.copy()
# La columna 'vaccines' puede contener valores nulos. Los tratamos como 0 vacunas.
conCantidad['canVac'] = conCantidad['vaccines'].apply(lambda x: len(str(x).split(', ')) if pd.notna(x) else 0)

display(conCantidad.head())
```



---

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.


---



```python
antes20 = country_vaccinations[country_vaccinations['date'] < pd.to_datetime('2020-12-20')]

print("DataFrame antes20 (registros antes del 20 de diciembre de 2020):")
display(antes20.head())
print("\nInformación del DataFrame antes20:")
antes20.info()
```



---

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer


---



```python
pfizer = country_vaccinations[country_vaccinations['vaccines'].str.contains('Pfizer', na=False)]
display(pfizer.head())
print(f"Número de registros donde se utilizó Pfizer: {len(pfizer)} de {len(country_vaccinations)}")
```



---


l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.


---



```python
!pip install xlsxwriter
```

```python
output_file = 'resultadosReto.xlsx'

with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
    country_vaccinations_sorted.to_excel(writer, sheet_name='country_vaccinations_sorted', index=False)

print(f"Dataframes guardados exitosamente en '{output_file}'")
```
